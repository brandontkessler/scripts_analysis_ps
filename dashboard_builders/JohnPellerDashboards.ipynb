```python
from transform_ps.transform_ps.donors import Donors
from transform_ps.transform_ps.tickets import Tickets
from transform_ps.transform_ps.subscriptions import Subscriptions

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Functions

```python
def filter_by_dt(df, dt):
    mask = df['perf_dt'] < dt
    df = df[mask].reset_index(drop=True)
    return df

def filt_by_dt_cont(df, dt):
    mask = df['cont_dt'] < dt
    df = df[mask].reset_index(drop=True)
    return df

def filt_by_dt_order(df, dt):
    mask = df['order_dt'] < dt
    df = df[mask].reset_index(drop=True)
    return df



def rev_tix(df):
    return round(sum(df['paid_amt']), 0)

def ct_tix(df):
    return len(df['paid_amt'])

def cust_tix(df):
    return len(set(df['summary_cust_id'])) 

def rev_don(df):
    return sum(df['gift_plus_pledge'])

def ct_don(df):
    return len(df['gift_plus_pledge'])

def cust_don(df):
    return len(set(df['summary_cust_id']))


def rev_sub(df):
    return round(sum(df['tot_due_amt']))

def tot_pkgs(df):
    return round(sum(df['num_seats']))



def plt_dict(d, title, ylab, xlab):
    keys = list(d.keys())
    vals = list(d.values())
    
    fig, ax = plt.subplots(figsize=(15,10))
    sns.set(style="ticks")

    # Create the plot
    plt.bar(keys, vals)

    # titles
    ax.set_title(title, fontsize=30)

    # axes
    ax.tick_params(axis='both', labelsize=16)
    ax.set_xlabel(xlab, fontsize=20)
    ax.set_ylabel(ylab, fontsize=20)

    plt.xticks(rotation=0)
    # Show plot
    plt.show()
```
